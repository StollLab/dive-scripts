```python
import numpy as np
import matplotlib.pyplot as plt
import dive 
import deerlab as dl
```

```python
loaded_data = np.genfromtxt("../../data/DEER_nonident_compact.txt", skip_header = 0, delimiter = ' ')
t = loaded_data[0,:]
V = loaded_data[1,:]
plt.plot(t,V)
plt.xlabel('t (µs)')
plt.ylabel('V (arb.u.)')
plt.show()
```

```python
pars = {"method": 'regularization', "r": np.linspace(2,8,200)}

model = dive.model(t, V, pars)

MCMCparameters = {"draws": 2000,"tune": 500, "chains": 4, 'cores': 4}
trace = dive.sample(model,MCMCparameters)
dive.summary(trace, model)  
```
