# Plotting

## Figure 2 - MCMC example

### First step 

```python
k_1 = trace_3992good_rmaxauto['k'][0]
lamb_1 = trace_3992good_rmaxauto['lamb'][0]
k_2 = trace_3992good_rmaxauto['k'][15]
lamb_2 = trace_3992good_rmaxauto['lamb'][15]
plt.scatter(k_1,lamb_1,s=50,c='k')
plt.scatter(k_2,lamb_2,s=50,c='#F38D68')
plt.ylabel(r'$\lambda$')
plt.ylim([0.5, 0.58])
plt.xlim([0.0, 0.058])
plt.xlabel(r'k ($\mu s^{-1}$)')
plt.savefig("step1.svg")
```

### First 25 steps

```python
import random
nChainSamples = trace_3992good_rmaxauto['P'].shape[0]
nDraws= 25
idxSamples = random.sample(range(nChainSamples), nDraws)
k = trace_3992good_rmaxauto['k'][idxSamples]
lamb = trace_3992good_rmaxauto['lamb'][idxSamples]
k_25 = k[24]
lamb_25 = lamb[24]
k_26 = trace_3992good_rmaxauto['k'][idxSamples[24]+1]
lamb_26 = trace_3992good_rmaxauto['lamb'][idxSamples[24]+1]

plt.scatter(k,lamb,s=50,c='k',alpha=0.5)
plt.scatter(k_25,lamb_25,s=50,c='k')
plt.scatter(k_26,lamb_26,s=50,c='#F38D68')
plt.ylabel(r'$\lambda$')
plt.ylim([0.5, 0.58])
plt.xlim([0.0, 0.058])
plt.xlabel(r'k ($\mu s^{-1}$)')
plt.savefig("steps1_25.svg")
```

### Chains 1 & 2 (partial)

```python
import random
nChainSamples_c1_l = 20000
nChainSamples_c1_u= 39999
nChainSamples_c2_l = 60000
nChainSamples_c2_u = 79999
nDraws= 1000
idxSamples_c1 = random.sample(range(nChainSamples_c1_l,nChainSamples_c1_u), nDraws)
k_c1 = trace_3992good_rmaxauto['k'][idxSamples_c1]
lamb_c1 = trace_3992good_rmaxauto['lamb'][idxSamples_c1]
idxSamples_c2 = random.sample(range(nChainSamples_c2_l,nChainSamples_c2_u), nDraws)
k_c2 = trace_3992good_rmaxauto['k'][idxSamples_c2]
lamb_c2 = trace_3992good_rmaxauto['lamb'][idxSamples_c2]

plt.scatter(k_c1,lamb_c1,s=50,c='k',alpha=0.5)
plt.scatter(k_c2,lamb_c2,s=50,c='#0B6E4F',alpha=0.5)
plt.ylabel(r'$\lambda$')
plt.ylim([0.5, 0.58])
plt.xlim([0.0, 0.058])
plt.xlabel(r'k ($\mu s^{-1}$)')
plt.savefig("chains1_2.svg")
```

### Final posteriors

```python
ax = az.plot_pair(trace_3992good_rmaxauto, var_names=['k','lamb'], kind='kde')
ax.set_xlim([0, 0.058])
ax.set_ylim([0.5, 0.58])
fig = plt.gcf() # to get the current figure...
fig.savefig("klamb_posterior.svg") # and save it directly
```

```python
ax1 = az.plot_kde(trace_3992good_rmaxauto['k'])
ax1.set_xlim([0, 0.058])
fig = plt.gcf() # to get the current figure...
fig.savefig("k_posterior.svg") # and save it directly
```

```python
ax2 = az.plot_kde(trace_3992good_rmaxauto['lamb'])
ax2.set_xlim([0.5, 0.58])
fig = plt.gcf() # to get the current figure...
fig.savefig("lamb_posterior.svg") # and save it directly
```
